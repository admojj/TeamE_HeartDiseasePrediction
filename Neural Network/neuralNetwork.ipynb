```python
import pandas as pd    #import several important packages which give us some useful features
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt

#import the library classes for the neural network model
from keras.models import Sequential
from keras.layers import Dense

from numpy import array
```

```python
clevelandData = pd.read_csv('processed.cleveland1.csv', delimiter=',') #loading data
clevelandData
```

```python
clevelandData.drop(columns = ['trestbps', 'fbs', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal' ], inplace=True, errors= 'ignore')
clevelandData
```

```python
clevelandData.columns = ['Age', 'Sex', 'CP', 'Chol', 'ECG', 'class']
clevelandData
```

```python
clevelandData["Class"] = clevelandData["class"]
clevelandData
```

```python
clevelandData['Class'] = ['0' if x==0 else '1' for x in clevelandData['class']]
clevelandData
```

```python
clevelandData.drop(columns = ['class'], inplace = True, errors = 'ignore')
clevelandData
```

```python
clevelandData.isnull().sum() #compute sum of missing values per column
```

```python
#Normalisation
x = clevelandData.values
```

```python
print (x)
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
clevelandData = clevelandData.copy()
clevelandDatanorm = pd.DataFrame(x_scaled, columns=clevelandData.columns)
```

```python
clevelandDatanorm.head(len(clevelandDatanorm))
```

```python
#Here starts the neural network computation

# split into input (X) and output (y) variables

clevelandDatanorm.to_csv('processed.cleveland2.csv', header=False)
```

```python
clevelandDatanormArray = np.loadtxt('processed.cleveland2.csv', delimiter=',')
```

```python
X = clevelandDatanormArray[:,0:5]
y = clevelandDatanormArray[:,5]
```

```python
# define the keras model

model = Sequential()

model.add(Dense(2000, input_dim=5, activation='relu'))
model.add(Dense(1000, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset

model.fit(X, y, epochs=1000, batch_size=20)
```

```python
# evaluate the keras model

_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
